### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt

# Діапазон значень x (години)
x = np.linspace(0, 10, 100)

# Функція оцінки
def f(x):
    return 10 * x - 5

# Похідна функції
def f_prime(x):
    return 10 * np.ones_like(x)

# Обчислення значень функції та похідної
y = f(x)
y_prime = f_prime(x)

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Оцінка f(x) = 10x - 5', color='blue')
plt.plot(x, y_prime, label='Похідна f\'(x) = 10', color='red', linestyle='--')
plt.title('Графік функції та її похідної')
plt.xlabel('Години навчання')
plt.ylabel('Оцінка / Зміна оцінки')
plt.legend()
plt.grid(True)
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

# Оголошуємо символічну змінну
x = sp.Symbol('x')

# Функції
f1 = sp.sin(x)
f2 = sp.ln(x)

# Обчислюємо похідні
f1_derivative = sp.diff(f1, x)
f2_derivative = sp.diff(f2, x)

# Виводимо аналітичні похідні
print(f"Похідна sin(x): {f1_derivative}")
print(f"Похідна ln(x): {f2_derivative}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Створюємо діапазон значень
x_vals1 = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x_vals2 = np.linspace(0.1, 10, 400)

# Функції та похідні
f1_vals = np.sin(x_vals1)
f1_deriv_vals = np.cos(x_vals1)

f2_vals = np.log(x_vals2)
f2_deriv_vals = 1 / x_vals2

# Створення графіків
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# sin(x)
axs[0].plot(x_vals1, f1_vals, label='f(x) = sin(x)', color='blue')
axs[0].plot(x_vals1, f1_deriv_vals, label="f'(x) = cos(x)", color='red', linestyle='--')
axs[0].set_title('sin(x) та похідна')
axs[0].legend()
axs[0].grid(True)

# ln(x)
axs[1].plot(x_vals2, f2_vals, label='f(x) = ln(x)', color='green')
axs[1].plot(x_vals2, f2_deriv_vals, label="f'(x) = 1/x", color='orange', linestyle='--')
axs[1].set_title('ln(x) та похідна')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Задаємо змінну
x = sp.Symbol('x')

# Функція витрат
f = x**2 - 5*x + 1

# Похідна функції
f_derivative = sp.diff(f, x)

# Вивід результатів
print(f"Функція витрат: {f}")
print(f"Похідна функції: {f_derivative}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Діапазон значень x
x_vals = np.linspace(0, 100, 400)

# Обчислення значень функції та похідної
f_vals = x_vals**2 - 5 * x_vals + 1
f_deriv_vals = 2 * x_vals - 5

# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = x² - 5x + 1', color='blue')
plt.plot(x_vals, f_deriv_vals, label="f'(x) = 2x - 5", color='red', linestyle='--')
plt.title('Функція загальних витрат та її похідна')
plt.xlabel('Кількість одиниць продукції (x)')
plt.ylabel('Значення')
plt.grid(True)
plt.legend()
plt.show()
```
